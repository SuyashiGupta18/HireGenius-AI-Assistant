```python
def clean_resume(text):

    text = str(text)

    # remove urls
    text = re.sub(r'http\S+|www\S+', ' ', text)

    # remove email
    text = re.sub(r'\S+@\S+', ' ', text)

    # remove phone numbers
    text = re.sub(r'\d{10,}', ' ', text)

    # remove special characters
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    # lowercase
    text = text.lower()

    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
def calculate_skill_match(candidate_skills,
                          required_skills):

    candidate_set = set(candidate_skills)

    required_set = set(required_skills)

    matched = candidate_set.intersection(required_set)

    score = (
        len(matched) /
        len(required_set)
    ) * 100

    return round(score,2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def calculate_resume_similarity(
    resume_text,
    job_description
):

    documents = [
        str(resume_text),
        str(job_description)
    ]

    tfidf = TfidfVectorizer()

    matrix = tfidf.fit_transform(documents)

    similarity = cosine_similarity(
        matrix[0:1],
        matrix[1:2]
    )

    return round(
        similarity[0][0] * 100,
        2
    )
```

```python
def generate_explanation(candidate_skills,
                         required_skills,
                         similarity_score,
                         final_score):

    candidate_set = set(candidate_skills)

    required_set = set(required_skills)

    matched_skills = list(
        candidate_set.intersection(required_set)
    )

    missing_skills = list(
        required_set - candidate_set
    )

    recommendation = hiring_recommendation(
        final_score
    )

    return {
        "Matched Skills": matched_skills,
        "Missing Skills": missing_skills,
        "Similarity Score": similarity_score,
        "Final Score": final_score,
        "Recommendation": recommendation
    }
```

```python
def extract_text_from_pdf(pdf_path):

    text = ""

    with pdfplumber.open(pdf_path) as pdf:

        for page in pdf.pages:

            page_text = page.extract_text()

            if page_text:
                text += page_text + "\n"

    return text
```

```python
def extract_text_from_docx(docx_path):

    doc = docx.Document(docx_path)

    text = ""

    for para in doc.paragraphs:

        text += para.text + "\n"

    return text
```

```python
def generate_hiring_report_v5():

    pdf = SimpleDocTemplate(
        "HireGenius_Professional_Report_V5.pdf"
    )

    styles = getSampleStyleSheet()

    title_style = ParagraphStyle(
        "TitleStyle",
        parent=styles["Title"],
        textColor=colors.HexColor("#131414"),
        alignment=1,
        fontSize=24
    )

    section_style = ParagraphStyle(
        "SectionStyle",
        parent=styles["Heading2"],
        textColor=colors.HexColor("#131414")
    )

    elements = []

    # ======================================
    # HEADER
    # ======================================

    subtitle_style = ParagraphStyle(
    "SubtitleStyle",
    parent=styles["Normal"],
    alignment=1,
    fontSize=12,
    textColor=colors.grey
)


    elements.append(
    Paragraph(
        "HIREGENIUS AI",
        title_style
    )
)

    elements.append(
    Paragraph(
        "AI-Powered Candidate Assessment Report",
        subtitle_style
    )
)

    elements.append(
    Spacer(1,8)
)

    elements.append(
        Spacer(1,15)
    )

    elements.append(
        HRFlowable(
            width="100%",
            thickness=1,
            color=colors.grey
        )
    )

    elements.append(
        Spacer(1,15)
    )

    # ======================================
    # CANDIDATE OVERVIEW
    # ======================================

    elements.append(
        Paragraph(
            "Candidate Overview",
            section_style
        )
    )

    overview = [

        ["Candidate Name", candidate_name],

        ["Applied Role", job_role],

        ["Report Date",
         datetime.now().strftime("%d-%m-%Y")]

    ]

    overview_table = Table(
        overview,
        colWidths=[180,220]
    )

    overview_table.setStyle(
        TableStyle([

            ('GRID',(0,0),(-1,-1),1,colors.grey),

            ('BACKGROUND',
             (0,0),
             (-1,-1),
             colors.whitesmoke)

        ])
    )

    elements.append(
        overview_table
    )

    elements.append(
        Spacer(1,20)
    )

    # ======================================
    # PERFORMANCE DASHBOARD
    # ======================================

    score_card = Table(



    [[
        f"""
    Final Score

    {round(final_score,2)}%

    {recommendation.upper()}
    """
        ]],

        colWidths=[400]

    )

    score_card.setStyle(

        TableStyle([

            ('BACKGROUND',
            (0,0),
            (-1,-1),
            colors.HexColor("#b6fcbf")),

            ('ALIGN',
            (0,0),
            (-1,-1),
            'CENTER'),

            ('BOX',
            (0,0),
            (-1,-1),
            2,
            colors.HexColor("#4e8a5d"))

        ])

    )

    elements.append(
        score_card
    )

    elements.append(
        Spacer(1,20)
    )

    elements.append(
        Paragraph(
            "Performance Dashboard",
            section_style
        )
    )

    rec_text = (
    "🟢 RECOMMEND"
    if recommendation.lower() == "recommend"
    else "🔴 REJECT"
)

    dashboard = [

        ["Metric","Score"],

        ["Skill Match",
         f"{round(skill_match,2)}%"],

        ["Similarity",
         f"{round(similarity,2)}%"],

        ["Final Score",
         f"{round(final_score,2)}%"],

        ["Recommendation", rec_text]

    ]

    dashboard_table = Table(
        dashboard,
        colWidths=[200,200]
    )

    dashboard_table.setStyle(
        TableStyle([

            ('BACKGROUND',
             (0,0),
             (-1,0),
             colors.HexColor("#322e36")),

            ('TEXTCOLOR',
             (0,0),
             (-1,0),
             colors.white),

            ('GRID',
             (0,0),
             (-1,-1),
             1,
             colors.black),

            ('ALIGN',
             (0,0),
             (-1,-1),
             'CENTER')

        ])
    )

    elements.append(
        dashboard_table
    )

    elements.append(
    Spacer(1,15)
)

    elements.append(
        Paragraph(
            f"Skill Match: {create_progress_bar(skill_match)} {round(skill_match,2)}%",
            styles["BodyText"]
        )
    )

    elements.append(
        Paragraph(
            f"Similarity: {create_progress_bar(similarity)} {round(similarity,2)}%",
            styles["BodyText"]
        )
    )

    elements.append(
        Paragraph(
            f"Final Score: {create_progress_bar(final_score)} {round(final_score,2)}%",
            styles["BodyText"]
        )
    )

    elements.append(
        Spacer(1,20)
    )

    # ======================================
    # DETECTED SKILLS
    # ======================================

    elements.append(
        Paragraph(
            "Detected Skills",
            section_style
        )
    )

    skill_rows = [

    candidate_skills[i:i+3]

    for i in range(
        0,
        len(candidate_skills),
        3
    )

]

    skills_table = Table(
        skill_rows,
        colWidths=[120,120,120]
    )

    skills_table.setStyle(
        TableStyle([

            ('BACKGROUND',
            (0,0),
            (-1,-1),
           colors.HexColor("#d7dbd7")),

            ('GRID',
            (0,0),
            (-1,-1),
            1,
            colors.white),

            ('ALIGN',
            (0,0),
            (-1,-1),
            'CENTER')

        ])
    )

    elements.append(
        skills_table
    )
    elements.append(
            Spacer(1,15)
        )

    # ======================================
    # STRENGTHS
    # ======================================

    elements.append(
        Paragraph(
            "Strength Analysis",
            section_style
        )
    )

    strengths = get_strengths(
        candidate_skills
    )

    for s in strengths:

        elements.append(
            Paragraph(
                f"✓ {s}",
                styles["BodyText"]
            )
        )

    elements.append(
        Spacer(1,15)
    )

    # ======================================
    # MISSING SKILLS
    # ======================================

    elements.append(
        Paragraph(
            "Skill Gap Analysis",
            section_style
        )
    )

    missing = get_missing_skills(
        candidate_skills,
        required_skills
    )

    if len(missing) == 0:

        elements.append(
            Paragraph(
                "No critical skill gaps detected.",
                styles["BodyText"]
            )
        )

    else:

        for skill in missing:

            elements.append(
                Paragraph(
                    f"✗ {skill}",
                    styles["BodyText"]
                )
            )

    elements.append(
        Spacer(1,15)
    )



    # ======================================
    # AI VERDICT
    # ======================================



    elements.append(
        Paragraph(
            "AI Recruiter Verdict",
            section_style
        )
    )

    elements.append(
        Paragraph(
            ai_summary(),
            styles["BodyText"]
        )
    )

    elements.append(
        Spacer(1,10)
    )

    recruiter_note = """
    <b>Recruiter Recommendation</b><br/><br/>

    • Candidate demonstrates strong alignment with the target role.<br/>
    • Relevant technical skills have been identified successfully.<br/>
    • Overall score indicates good hiring potential.<br/>
    • Recommended for the next stage of evaluation or technical interview.
    """

    note_data = [[
        Paragraph(
            recruiter_note,
            styles["BodyText"]
        )
    ]]

    note_table = Table(
        note_data,
        colWidths=[400]
    )

    note_table.setStyle(
        TableStyle([
            ('BACKGROUND',
             (0,0),
             (-1,-1),
             colors.whitesmoke),

            ('BOX',
             (0,0),
             (-1,-1),
             1,
             colors.grey),

            ('LEFTPADDING',
             (0,0),
             (-1,-1),
             10),

            ('RIGHTPADDING',
             (0,0),
             (-1,-1),
             10),

            ('TOPPADDING',
             (0,0),
             (-1,-1),
             10),

            ('BOTTOMPADDING',
             (0,0),
             (-1,-1),
             10)
        ])
    )

    elements.append(
        note_table
    )

    elements.append(
        Spacer(1,15)
    )

    # ======================================
    # INTERVIEW QUESTIONS
    # ======================================

    elements.append(
        Paragraph(
            "Recommended Interview Questions",
            section_style
        )
    )

    questions = generate_interview_questions(
        candidate_skills
    )

    for q in questions:

        elements.append(
            Paragraph(
                f"• {q}",
                styles["BodyText"]
            )
        )

    elements.append(
        Spacer(1,20)
    )

    # ======================================
    # FOOTER
    # ======================================

    elements.append(
        HRFlowable(
            width="100%",
            thickness=1,
            color=colors.grey
        )
    )

    elements.append(
        Spacer(1,10)
    )

    elements.append(
        Paragraph(
            "Generated by HireGenius AI Recruitment Platform",
            styles["Italic"]
        )
    )

    pdf.build(elements)

    print(
        "HireGenius V5 Report Generated Successfully"
    )
```

```python
def generate_resume_feedback(
    candidate_skills,
    required_skills,
    final_score
):

    feedback = []

    matched = list(
        set(candidate_skills)
        &
        set(required_skills)
    )

    missing = list(
        set(required_skills)
        -
        set(candidate_skills)
    )

    # ==========================
    # STRENGTHS
    # ==========================

    feedback.append("=== STRENGTHS ===")

    if len(matched):

        for skill in matched:

            feedback.append(
                f"✓ Strong in {skill}"
            )

    else:

        feedback.append(
            "No matching skills found."
        )

    feedback.append("")

    # ==========================
    # IMPROVEMENT AREAS
    # ==========================

    feedback.append(
        "=== IMPROVEMENT AREAS ==="
    )

    if len(missing):

        for skill in missing:

            feedback.append(
                f"✗ Learn {skill}"
            )

    else:

        feedback.append(
            "✓ No critical skill gaps detected."
        )

    feedback.append("")

    # ==========================
    # RECOMMENDATIONS
    # ==========================

    feedback.append(
        "=== RECOMMENDATIONS ==="
    )

    if final_score >= 80:

        feedback.append(
            "Candidate is highly aligned with the role."
        )

    elif final_score >= 60:

        feedback.append(
            "Candidate should strengthen missing skills."
        )

    else:

        feedback.append(
            "Candidate requires significant upskilling."
        )

    feedback.append(
        "Add real-world projects."
    )

    feedback.append(
        "Maintain GitHub portfolio."
    )

    return "\n".join(feedback)
```

```python
feedback = generate_resume_feedback(
    candidate_skills,
    required_skills,
    final_score
)

print(feedback)
```

```python
def recruiter_chatbot(question):

    q = question.lower()

    missing = list(
        set(required_skills)
        -
        set(candidate_skills)
    )

    if "score" in q:

        return f"""
Candidate Final Score: {round(final_score,2)}%
Recommendation: {recommendation}
"""

    elif "skills" in q:

        return (
            "Detected Skills:\n\n"
            +
            ", ".join(candidate_skills)
        )

    elif "missing" in q:

        if len(missing) == 0:

            return (
                "No critical skill gaps detected."
            )

        return (
            "Missing Skills:\n\n"
            +
            ", ".join(missing)
        )

    elif "why" in q:

        return f"""
Candidate possesses
{len(candidate_skills)} identified skills.

Skill Match: {round(skill_match,2)}%

Final Score: {round(final_score,2)}%

Recommendation:
{recommendation}
"""

    elif "interview" in q:

        questions = generate_interview_questions(
            candidate_skills
        )

        return "\n".join(questions)

    elif "feedback" in q:

        return generate_resume_feedback(
            candidate_skills,
            required_skills,
            final_score
        )

    else:

        return """
You can ask:

• score
• skills
• missing skills
• recommendation
• why recommended
• interview questions
• feedback
"""
```
